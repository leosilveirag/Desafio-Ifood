# **Análise de Correlação dos Dados**





### **Pergunta 1**
 Qual é a relação de estado civil com número de filhos? Será que as pessoas casadas têm um maior número de filhos?

```python
# Agrupa os dados por 'marital_status' e conta quantos clientes têm ou não filhos em cada grupo
df.groupby('marital_status')['kids'].value_counts()
```

```python
# fazendo um groupby para relacionar as duas colunas e fazendo uma media
df.groupby('marital_status')['kids'].mean().round(2)
```

```python
# fazendo um groupby para relacionar as duas colunas e fazendo uma mediana
df.groupby('marital_status')['kids'].median().round(2)
```

 ### **Resposta**
 <p align= "justify">Apesar de a média e a mediana de filhos serem consistentemente 1 para todos os estados civis, há algumas nuances importantes a serem consideradas:

-  **Evidência central**: A concentração de clientes com 1 filho é significativamente maior no grupo de clientes casados. Isso indica que, de fato, a maioria dos clientes casados se enquadra na categoria de ter um filho.
- **Pontos de atenção**: Por outro lado, o grupo de clientes casados também apresenta o maior número de famílias sem filhos (zero filhos), comparado a outros estados civis.

<p align= "justify">Em resumo, a tendência geral é que clientes casados tenham mais filhos, com a maioria das famílias tendo 1 filho. No entanto, o grupo também inclui uma grande parcela de casais sem filhos, o que mostra a diversidade dentro desse estado civil.

### **Pergunta 2:**
As pessoas gastam mais ou menos em nossa plataforma quando têm filhos?

```python
# usando a media para ver se os gastos são maiores com filhos
df.groupby('kids')['expenses'].median()
```

```python
# usando a media para ver se os gastos são maiores com filhos
df.groupby('kids')['expenses'].mean().round(2)
```

### **Resposta**
<p align= "justify">Com base na média e na mediana dos gastos, é possível notar uma clara relação inversa entre o número de filhos e o valor gasto na plataforma.

- **Gasto maior**: Clientes sem nenhum filho (0 filhos) apresentam uma média e uma mediana de gastos significativamente maiores do que os clientes com filhos.
- **Gasto menor**: À medida que o número de filhos aumenta, os gastos com a plataforma tendem a diminuir.

<p align= "justify">Portanto, as pessoas que não tem filhos gastam mais na plataforma do que as pessoas que tem filhos. E quanto mais filhos tem, menor o gasto.

```python
# realizando mediana dos valores das 100 menores rendas com os 100 menores gastos
amostra_menor_renda.median()
```

```python
# realizando mediana dos valores das 100 maiores rendas com os 100 maiores gastos
amostra_maior_renda.median()
```

```python
# realizando a média da amostra da menor renda e do menor gasto
amostra_menor_renda.mean()
```

```python
# realizando a média da amostra da maior renda e do maior gasto
amostra_maior_renda.mean()
```

```python
# resumo da estatística descritiva(ver especialmente o mínimo e máx)
amostra_menor_renda.describe()
```

```python
# resumo da estatística descritiva(ver especialmente o mínimo e máx)
amostra_maior_renda.describe()
```

 ### **Resposta e Processo Realizado**

<p align= "justify">Sim, no geral pessoas que tem uma renda maior gastam mais do que quem tem uma renda menor.
Ao realizar a comparação entre os grupos de clientes de maior e menor renda, usando métricas como a média e a mediana identificamos que:

- A média dos gastos dos clientes de maior renda é significativamente maior do que a média dos gastos dos clientes de menor renda.
- Já a mediana de gastos, que é um indicador mais robusto, também confirma essa tendência, mostrando um valor muito maior para o grupo de maior renda do que para o de menor renda.

<p align= "justify">Notamos também, uma possível presença de outlier no grupo de menor renda, um cliente que gastou um valor considerável que estava fugindo do padrão dos últimos cinco clientes de menor renda, mas que não foi suficiente para distorcer a conclusão geral da análise.

### **Processo para resultado final**
<p align= "justify">Inicialmente, tive a ideia de comparar as 5 menores e 5 maiores rendas do dataframe e depois de comparar os gastos. Aprimorei essa ideia para uma amostra maior e mais confiável: as 100 menores rendas e as 100 maiores rendas, ou seja, mais o menos 10% dos dados.
Depois, analisei os extremos e criei dois grupos distintos para análise, um contendo os 100 clientes com as menores rendas e os menores gastos, e outro com os 100 clientes com as maiores rendas e os maiores gastos.
